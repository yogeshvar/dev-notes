```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plotter
```

```python
dataset = pd.read_csv('datasets/Uber_data.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, : 1].values # get no of uber drivers
Y = dataset.iloc[:, 1].values # get average trip
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

## Simple Linear Regression

---

### Linear Regression Function:
<img src="https://latex.codecogs.com/gif.latex?y&space;=&space;a&space;&plus;&space;bx" title="y = a + bx" />

* a => y-intercept
* b => slope
* x => independent variable
* y => dependent variable

---

### Slope(b) of Regression Line:

<img src="https://latex.codecogs.com/gif.latex?b&space;=&space;r&space;\left&space;(&space;\frac{S_y}{S_x}&space;\right&space;)" title="b = r \left ( \frac{S_y}{S_x} \right )" />

* b => slope
* r => pearson correlation coefficient
* S.D => standard deviation

---

### Y-intercept(a) of Linear Regression:

<img src="https://latex.codecogs.com/gif.latex?a&space;=&space;\bar{y}&space;-&space;b\bar{x}" title="a = \bar{y} - b\bar{x}" />

* a => y-intercept
* <img src="https://latex.codecogs.com/gif.latex?\bar{y}" /> => mean value of Y
* <img src="https://latex.codecogs.com/gif.latex?\bar{x}" /> => mean value of X

---

### Pearson correlation coefficient:
<img src="https://latex.codecogs.com/gif.latex?r&space;=&space;\frac{\sum&space;\left&space;(&space;x&space;-&space;\bar{x}&space;\right&space;)&space;\left&space;(&space;y&space;-&space;\bar{y}&space;\right&space;)}{\sqrt{\sum&space;\left&space;(&space;x&space;-&space;\bar{x}&space;\right&space;)^{2}\left&space;(&space;y&space;-&space;\bar{y}&space;\right&space;)^{2}}}" title="r = \frac{\sum \left ( x - \bar{x} \right ) \left ( y - \bar{y} \right )}{\sqrt{\sum \left ( x - \bar{x} \right )^{2}\left ( y - \bar{y} \right )^{2}}}" />

### Standard Deviation:
